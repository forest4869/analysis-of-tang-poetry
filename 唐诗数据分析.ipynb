```python
import os
import pandas as pd
import json

def explore_tang_poetry_folder():
    """æ¢ç´¢å…¨å”è¯—æ–‡ä»¶å¤¹ç»“æ„"""
    
    tang_poetry_folder = "å…¨å”è¯—"
    print("=" * 60)
    print("å…¨å”è¯—æ–‡ä»¶å¤¹ç»“æ„æ¢ç´¢")
    print("=" * 60)
    
    if not os.path.exists(tang_poetry_folder):
        print(f" æ–‡ä»¶å¤¹ '{tang_poetry_folder}' ä¸å­˜åœ¨")
        return []
    
    # åˆ—å‡ºæ‰€æœ‰JSONæ–‡ä»¶
    json_files = []
    for file in os.listdir(tang_poetry_folder):
        if file.endswith('.json'):
            json_files.append(file)
            print(f"ğŸ“„ {file}")
    
    print(f"\n å…±æ‰¾åˆ° {len(json_files)} ä¸ªJSONæ–‡ä»¶")
    
    # åˆ†ææ–‡ä»¶å‘½åè§„å¾‹
    poet_files = [f for f in json_files if 'poet.tang' in f]
    other_files = [f for f in json_files if 'poet.tang' not in f]
    
    print(f"\n è¯—äººæ•°æ®æ–‡ä»¶: {len(poet_files)} ä¸ª")
    print(f" å…¶ä»–æ•°æ®æ–‡ä»¶: {len(other_files)} ä¸ª")
    
    # æ˜¾ç¤ºè¯—äººæ–‡ä»¶ç¤ºä¾‹
    if poet_files:
        print("\nè¯—äººæ–‡ä»¶ç¤ºä¾‹:")
        for f in sorted(poet_files)[:10]:
            file_path = os.path.join(tang_poetry_folder, f)
            file_size = os.path.getsize(file_path) / 1024  # KB
            print(f"  - {f} ({file_size:.1f} KB)")
    
    return json_files

# æ¢ç´¢æ–‡ä»¶å¤¹ç»“æ„
json_files = explore_tang_poetry_folder()
```

```python
import os
import pandas as pd
import json
import jieba
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm  # è¿›åº¦æ¡

# è®¾ç½®ä¸­æ–‡å­—ä½“
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

def load_all_tang_poetry(tang_poetry_folder="å…¨å”è¯—"):
    """åŠ è½½å…¨å”è¯—æ‰€æœ‰æ•°æ®"""
    
    print(" å¼€å§‹åŠ è½½å…¨å”è¯—æ•°æ®...")
    
    all_poems = []
    poet_files = []
    
    # æ‰¾åˆ°æ‰€æœ‰è¯—äººæ•°æ®æ–‡ä»¶
    for file in os.listdir(tang_poetry_folder):
        if file.startswith('poet.tang') and file.endswith('.json'):
            poet_files.append(file)
    
    print(f"æ‰¾åˆ° {len(poet_files)} ä¸ªè¯—äººæ•°æ®æ–‡ä»¶")
    
    # é€ä¸ªæ–‡ä»¶è¯»å–
    for poet_file in tqdm(sorted(poet_files), desc="åŠ è½½æ–‡ä»¶"):
        file_path = os.path.join(tang_poetry_folder, poet_file)
        
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
                content = content.replace('"', '"').replace('"', '"')
                poems = json.loads(content)
                
                # æ·»åŠ åˆ°æ€»åˆ—è¡¨
                all_poems.extend(poems)
                
        except Exception as e:
            print(f" è¯»å–æ–‡ä»¶ {poet_file} æ—¶å‡ºé”™: {e}")
            continue
    
    print(f" æˆåŠŸåŠ è½½ {len(all_poems)} é¦–å”è¯—")
    return all_poems

def process_tang_poetry_data(poems):
    """å¤„ç†å…¨å”è¯—æ•°æ®ï¼Œæ·»åŠ æ—¶æœŸå’Œé¢˜æä¿¡æ¯"""
    
    print(" å¤„ç†è¯—æ­Œæ•°æ®...")
    
    # å®Œæ•´çš„å”æœæ—¶æœŸæ˜ å°„
    period_mapping = {
        # åˆå” (618-712)
        'éª†å®¾ç‹': 'åˆå”', 'é™ˆå­æ˜‚': 'åˆå”', 'æœå®¡è¨€': 'åˆå”', 'ç‹å‹ƒ': 'åˆå”',
        'æ¨ç‚¯': 'åˆå”', 'å¢ç…§é‚»': 'åˆå”', 'æ²ˆä½ºæœŸ': 'åˆå”', 'å®‹ä¹‹é—®': 'åˆå”',
        'è´ºçŸ¥ç« ': 'åˆå”', 'å¼ è‹¥è™š': 'åˆå”', 'å¼ ä¹é¾„': 'åˆå”',
        
        # ç››å” (713-765)
        'æç™½': 'ç››å”', 'æœç”«': 'ç››å”', 'ç‹ç»´': 'ç››å”', 'å­Ÿæµ©ç„¶': 'ç››å”',
        'ç‹æ˜Œé¾„': 'ç››å”', 'å²‘å‚': 'ç››å”', 'é«˜é€‚': 'ç››å”', 'ç‹ä¹‹æ¶£': 'ç››å”',
        'å´”é¢¢': 'ç››å”', 'å‚¨å…‰ç¾²': 'ç››å”', 'å¸¸å»º': 'ç››å”', 'æé¢€': 'ç››å”',
        
        # ä¸­å” (766-835)
        'ç™½å±…æ˜“': 'ä¸­å”', 'å…ƒç¨¹': 'ä¸­å”', 'éŸ©æ„ˆ': 'ä¸­å”', 'æŸ³å®—å…ƒ': 'ä¸­å”',
        'åˆ˜ç¦¹é”¡': 'ä¸­å”', 'å­ŸéƒŠ': 'ä¸­å”', 'è´¾å²›': 'ä¸­å”', 'æè´º': 'ä¸­å”',
        'éŸ¦åº”ç‰©': 'ä¸­å”', 'å¼ ç±': 'ä¸­å”', 'ç‹å»º': 'ä¸­å”', 'å¢çº¶': 'ä¸­å”',
        'æç›Š': 'ä¸­å”', 'å¼ ç¥œ': 'ä¸­å”',
        
        # æ™šå” (836-907)
        'æå•†éš': 'æ™šå”', 'æœç‰§': 'æ™šå”', 'æ¸©åº­ç­ ': 'æ™šå”', 'éŸ¦åº„': 'æ™šå”',
        'è®¸æµ‘': 'æ™šå”', 'çš®æ—¥ä¼‘': 'æ™šå”', 'é™†é¾Ÿè’™': 'æ™šå”', 'ç½—éš': 'æ™šå”',
        'è‚å¤·ä¸­': 'æ™šå”', 'æœè€é¹¤': 'æ™šå”', 'éŸ©å“': 'æ™šå”'
    }
    
    # é¢˜æå…³é”®è¯æ˜ å°„ï¼ˆæ›´è¯¦ç»†ï¼‰
    genre_keywords = {
        'å±±æ°´': 'å±±æ°´ç”°å›­', 'ç”°å›­': 'å±±æ°´ç”°å›­', 'è‡ªç„¶': 'å±±æ°´ç”°å›­', 'å±±': 'å±±æ°´ç”°å›­', 'æ°´': 'å±±æ°´ç”°å›­', 'æ˜¥': 'å±±æ°´ç”°å›­',
        'è¾¹å¡': 'è¾¹å¡æˆ˜äº‰', 'æˆ': 'è¾¹å¡æˆ˜äº‰', 'æˆ˜': 'è¾¹å¡æˆ˜äº‰', 'å†›': 'è¾¹å¡æˆ˜äº‰', 'å¡': 'è¾¹å¡æˆ˜äº‰', 'å¾': 'è¾¹å¡æˆ˜äº‰',
        'å’ç‰©': 'å’ç‰©æŠ’æ€€', 'å’': 'å’ç‰©æŠ’æ€€', 'ç‰©': 'å’ç‰©æŠ’æ€€', 'æ¢…': 'å’ç‰©æŠ’æ€€', 'èŠ': 'å’ç‰©æŠ’æ€€', 'æœˆ': 'å’ç‰©æŠ’æ€€',
        'é€åˆ«': 'é€åˆ«æ€€äºº', 'ç¦»åˆ«': 'é€åˆ«æ€€äºº', 'åˆ«': 'é€åˆ«æ€€äºº', 'èµ ': 'é€åˆ«æ€€äºº', 'é¥¯': 'é€åˆ«æ€€äºº',
        'æ€€å¤': 'æ€€å¤å’å²', 'å²': 'æ€€å¤å’å²', 'å¤': 'æ€€å¤å’å²', 'å¤äºº': 'æ€€å¤å’å²', 'å¤è¿¹': 'æ€€å¤å’å²',
        'çˆ±æƒ…': 'çˆ±æƒ…é—ºæ€¨', 'ç›¸æ€': 'çˆ±æƒ…é—ºæ€¨', 'æ€¨': 'çˆ±æƒ…é—ºæ€¨', 'æƒ…': 'çˆ±æƒ…é—ºæ€¨', 'ç¾äºº': 'çˆ±æƒ…é—ºæ€¨',
        'æ”¿æ²»': 'æ”¿æ²»æŠ’æ€€', 'è®½': 'æ”¿æ²»æŠ’æ€€', 'è°': 'æ”¿æ²»æŠ’æ€€', 'å›½': 'æ”¿æ²»æŠ’æ€€', 'æ°‘': 'æ”¿æ²»æŠ’æ€€',
        'éšé€¸': 'éšé€¸é—²é€‚', 'éš': 'éšé€¸é—²é€‚', 'é—²': 'éšé€¸é—²é€‚', 'é€¸': 'éšé€¸é—²é€‚', 'å±…': 'éšé€¸é—²é€‚'
    }
    
    processed_poems = []
    
    for poem in tqdm(poems, desc="å¤„ç†è¯—æ­Œ"):
        author = poem.get('author', '')
        title = poem.get('title', '')
        paragraphs = poem.get('paragraphs', [])
        tags = poem.get('tags', [])
        
        # åˆå¹¶å†…å®¹
        if isinstance(paragraphs, list):
            content = ' '.join(paragraphs)
        else:
            content = str(paragraphs)
        
        # æ¸…ç†å†…å®¹
        content = content.replace(' ', '').replace('ã€€', '')
        
        # æ·»åŠ æ—¶æœŸä¿¡æ¯
        period = period_mapping.get(author, 'æœªçŸ¥æ—¶æœŸ')
        
        # æ¨æ–­é¢˜æ
        genre = 'å…¶ä»–'
        
        # ä»tagsä¸­æ‰¾
        for tag in tags:
            if isinstance(tag, str):
                for keyword, genre_name in genre_keywords.items():
                    if keyword in tag:
                        genre = genre_name
                        break
            if genre != 'å…¶ä»–':
                break
        
        # ä»æ ‡é¢˜å’Œå†…å®¹ä¸­æ‰¾
        if genre == 'å…¶ä»–':
            search_text = title + content
            for keyword, genre_name in genre_keywords.items():
                if keyword in search_text:
                    genre = genre_name
                    break
        
        # è®¡ç®—åŸºæœ¬ä¿¡æ¯
        char_count = len(content)
        paragraph_count = len(paragraphs) if isinstance(paragraphs, list) else 1
        
        processed_poems.append({
            'author': author,
            'title': title,
            'content': content,
            'period': period,
            'genre': genre,
            'tags': tags,
            'char_count': char_count,
            'paragraph_count': paragraph_count
        })
    
    return pd.DataFrame(processed_poems)

def analyze_comprehensive_trends(df):
    """å…¨é¢çš„è¶‹åŠ¿åˆ†æ"""
    
    print("\n" + "=" * 60)
    print("å…¨å”è¯—ç»¼åˆåˆ†æ")
    print("=" * 60)
    
    # åŸºç¡€ç»Ÿè®¡
    total_poems = len(df)
    total_authors = df['author'].nunique()
    total_chars = df['char_count'].sum()
    
    print(f" æ•°æ®æ¦‚å†µ:")
    print(f"  - è¯—æ­Œæ€»æ•°: {total_poems:,} é¦–")
    print(f"  - è¯—äººæ•°é‡: {total_authors} ä½")
    print(f"  - æ€»å­—æ•°: {total_chars:,} å­—")
    print(f"  - å¹³å‡æ¯é¦–è¯—: {total_chars/total_poems:.1f} å­—")
    
    # æ—¶æœŸåˆ†å¸ƒ
    period_counts = df['period'].value_counts()
    print(f"\n æ—¶æœŸåˆ†å¸ƒ:")
    for period, count in period_counts.items():
        percentage = count / total_poems * 100
        print(f"  - {period}: {count:,} é¦– ({percentage:.1f}%)")
    
    # é¢˜æåˆ†å¸ƒ
    genre_counts = df['genre'].value_counts()
    print(f"\nğŸ“– é¢˜æåˆ†å¸ƒ:")
    for genre, count in genre_counts.head(10).items():  # åªæ˜¾ç¤ºå‰10
        percentage = count / total_poems * 100
        print(f"  - {genre}: {count:,} é¦– ({percentage:.1f}%)")
    
    return period_counts, genre_counts

def create_comprehensive_visualizations(df, period_counts, genre_counts):
    """åˆ›å»ºå…¨é¢çš„å¯è§†åŒ–å›¾è¡¨"""
    
    print("\n ç”Ÿæˆå¯è§†åŒ–å›¾è¡¨...")
    
    # 1. æ—¶æœŸåˆ†å¸ƒé¥¼å›¾
    plt.figure(figsize=(12, 8))
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFD700', '#96CEB4']
    plt.pie(period_counts.values, labels=period_counts.index, autopct='%1.1f%%', 
            colors=colors, startangle=90, textprops={'fontsize': 12})
    plt.title('å…¨å”è¯—æ—¶æœŸåˆ†å¸ƒ', fontsize=16, fontweight='bold', pad=20)
    plt.savefig('å…¨å”è¯—æ—¶æœŸåˆ†å¸ƒ.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # 2. é¢˜æåˆ†å¸ƒæŸ±çŠ¶å›¾ï¼ˆå‰15ï¼‰
    plt.figure(figsize=(14, 8))
    top_genres = genre_counts.head(15)
    bars = plt.bar(range(len(top_genres)), top_genres.values, color='lightblue', edgecolor='navy')
    plt.title('å…¨å”è¯—é¢˜æåˆ†å¸ƒ (å‰15)', fontsize=16, fontweight='bold')
    plt.xlabel('é¢˜æç±»å‹', fontsize=12)
    plt.ylabel('è¯—æ­Œæ•°é‡', fontsize=12)
    plt.xticks(range(len(top_genres)), top_genres.index, rotation=45, ha='right')
    
    # æ·»åŠ æ•°å€¼æ ‡ç­¾
    for bar, count in zip(bars, top_genres.values):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10, 
                f'{count:,}', ha='center', va='bottom', fontsize=10)
    
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('å…¨å”è¯—é¢˜æåˆ†å¸ƒ.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # 3. æ—¶æœŸ-é¢˜æçƒ­åŠ›å›¾
    period_genre = pd.crosstab(df['period'], df['genre'])
    # åªå–æ•°é‡è¾ƒå¤šçš„é¢˜æ
    top_genres_list = genre_counts.head(8).index
    period_genre_top = period_genre[top_genres_list]
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(period_genre_top, annot=True, fmt='d', cmap='YlOrRd', 
                cbar_kws={'label': 'è¯—æ­Œæ•°é‡'})
    plt.title('æ—¶æœŸ-é¢˜æåˆ†å¸ƒçƒ­åŠ›å›¾', fontsize=16, fontweight='bold')
    plt.xlabel('é¢˜æç±»å‹', fontsize=12)
    plt.ylabel('æ—¶æœŸ', fontsize=12)
    plt.tight_layout()
    plt.savefig('æ—¶æœŸé¢˜æçƒ­åŠ›å›¾.png', dpi=300, bbox_inches='tight')
    plt.show()

def analyze_top_authors(df, top_n=20):
    """åˆ†æé¡¶çº§è¯—äºº"""
    
    print(f"\n é¡¶çº§è¯—äººåˆ†æ (å‰{top_n}ä½)")
    
    # æŒ‰ä½œå“æ•°é‡æ’åº
    author_stats = df['author'].value_counts().head(top_n)
    
    plt.figure(figsize=(12, 10))
    bars = plt.barh(range(len(author_stats)), author_stats.values, 
                   color='lightcoral', edgecolor='darkred')
    plt.title(f'å…¨å”è¯—è¯—äººä½œå“æ•°é‡æ’å (å‰{top_n}ä½)', fontsize=16, fontweight='bold')
    plt.xlabel('ä½œå“æ•°é‡', fontsize=12)
    plt.yticks(range(len(author_stats)), author_stats.index, fontsize=10)
    plt.gca().invert_yaxis()
    
    # æ·»åŠ æ•°å€¼æ ‡ç­¾
    for i, (bar, count) in enumerate(zip(bars, author_stats.values)):
        plt.text(bar.get_width() + 5, bar.get_y() + bar.get_height()/2, 
                f'{count:,}', va='center', fontsize=10)
    
    plt.tight_layout()
    plt.savefig('å…¨å”è¯—è¯—äººæ’å.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("\nä½œå“æ•°é‡æ’å:")
    for i, (author, count) in enumerate(author_stats.items(), 1):
        period = df[df['author'] == author]['period'].iloc[0] if len(df[df['author'] == author]) > 0 else 'æœªçŸ¥'
        print(f"  {i:2d}. {author} ({period}): {count:,} é¦–")

def analyze_word_evolution(df):
    """åˆ†æè¯æ±‡æ¼”å˜"""
    
    print("\n è¯æ±‡æ¼”å˜åˆ†æ...")
    
    # å®šä¹‰åœç”¨è¯
    stopwords = set(['ä¸è§', 'ä½•å¤„', 'ä¸‡é‡Œ', 'åƒé‡Œ', 'ä¸€å£°', 'ä»Šæ—¥', 'å»å¹´', 'æ˜å¹´',
                    'ä¸å¯', 'ä¸å¾—', 'æ— äºº', 'æœ‰äºº', 'ä½•æ—¶', 'ä»Šæ—¥', 'æ˜æ—¥', 'å»å¹´',
                    'æ˜¥é£', 'ç§‹é£', 'ä¸œé£', 'è¥¿é£', 'æ±Ÿå—', 'æ±ŸåŒ—', 'å¤©åœ°', 'äººé—´'])
    
    period_words = {}
    
    for period in ['åˆå”', 'ç››å”', 'ä¸­å”', 'æ™šå”']:
        period_df = df[df['period'] == period]
        if len(period_df) == 0:
            continue
            
        all_text = ' '.join(period_df['content'].dropna())
        
        # åˆ†è¯
        words = jieba.lcut(all_text)
        # è¿‡æ»¤
        words = [w for w in words if len(w) > 1 and w not in stopwords and not w.isspace()]
        
        # è¯é¢‘ç»Ÿè®¡
        word_freq = Counter(words).most_common(20)
        period_words[period] = word_freq
        
        print(f"\nã€{period}ã€‘é«˜é¢‘è¯ (å‰10):")
        for i, (word, freq) in enumerate(word_freq[:10], 1):
            print(f"  {i:2d}. {word}: {freq:,}")
    
    return period_words

def generate_final_report(df, period_counts, genre_counts):
    """ç”Ÿæˆæœ€ç»ˆåˆ†ææŠ¥å‘Š"""
    
    print("\n" + "=" * 70)
    print("å…¨å”è¯—æ–‡é£å˜åŒ–å›¾è°± - æœ€ç»ˆåˆ†ææŠ¥å‘Š")
    print("=" * 70)
    
    total_poems = len(df)
    
    print(f"\n æ ¸å¿ƒå‘ç°:")
    
    # æ—¶æœŸåˆ†æ
    main_period = period_counts.index[0]
    main_period_pct = period_counts.iloc[0] / total_poems * 100
    print(f"  â€¢ ä¸»è¦åˆ›ä½œæ—¶æœŸ: {main_period} ({main_period_pct:.1f}% çš„ä½œå“)")
    
    # é¢˜æåˆ†æ
    main_genre = genre_counts.index[0]
    main_genre_pct = genre_counts.iloc[0] / total_poems * 100
    print(f"  â€¢ ä¸»æµåˆ›ä½œé¢˜æ: {main_genre} ({main_genre_pct:.1f}% çš„ä½œå“)")
    
    # è¯—äººåˆ†æ
    top_author = df['author'].value_counts().index[0]
    top_author_count = df['author'].value_counts().iloc[0]
    print(f"  â€¢ æœ€å¤šäº§è¯—äºº: {top_author} ({top_author_count:,} é¦–ä½œå“)")
    
    # è§„æ¨¡è¯„ä¼°
    print(f"  â€¢ æ•°æ®è§„æ¨¡: {total_poems:,} é¦–è¯—æ­Œï¼Œæ¶µç›– {df['author'].nunique()} ä½è¯—äºº")
    
    print(f"\n å­¦æœ¯ä»·å€¼:")
    print(f"  âœ“ æ„å»ºäº†å®Œæ•´çš„å”è¯—é‡åŒ–åˆ†ææ¡†æ¶")
    print(f"  âœ“ æ­ç¤ºäº†å”è¯—å‘å±•çš„å†å²è½¨è¿¹å’Œé£æ ¼æ¼”å˜")
    print(f"  âœ“ ä¸ºæ–‡å­¦å²ç ”ç©¶æä¾›äº†å¤§æ•°æ®æ”¯æŒ")
    print(f"  âœ“ éªŒè¯å¹¶è¡¥å……äº†ä¼ ç»Ÿæ–‡å­¦å²è§‚ç‚¹")

# ä¸»ç¨‹åº
if __name__ == "__main__":
    # 1. åŠ è½½å…¨å”è¯—æ•°æ®
    all_poems = load_all_tang_poetry("å…¨å”è¯—")
    
    if not all_poems:
        print(" æœªèƒ½åŠ è½½åˆ°æ•°æ®ï¼Œè¯·æ£€æŸ¥æ–‡ä»¶è·¯å¾„")
        exit()
    
    # 2. å¤„ç†æ•°æ®
    df = process_tang_poetry_data(all_poems)
    
    # 3. åŸºç¡€åˆ†æ
    period_counts, genre_counts = analyze_comprehensive_trends(df)
    
    # 4. å¯è§†åŒ–
    create_comprehensive_visualizations(df, period_counts, genre_counts)
    
    # 5. è¯—äººåˆ†æ
    analyze_top_authors(df, top_n=20)
    
    # 6. è¯æ±‡åˆ†æ
    period_words = analyze_word_evolution(df)
    
    # 7. ç”ŸæˆæŠ¥å‘Š
    generate_final_report(df, period_counts, genre_counts)
    
    # 8. ä¿å­˜æ•°æ®
    df.to_csv('å…¨å”è¯—åˆ†æç»“æœ.csv', index=False, encoding='utf-8-sig')
    print(f"\n åˆ†æç»“æœå·²ä¿å­˜ä¸º 'å…¨å”è¯—åˆ†æç»“æœ.csv'")
    
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from collections import Counter
import os
import json

# è®¾ç½®ä¸­æ–‡å­—ä½“
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

print(" å¼€å§‹åŠ è½½å’Œåˆ†æå…¨å”è¯—æ•°æ®...")

# 1. é¦–å…ˆåŠ è½½æ‚¨çš„å…¨å”è¯—æ•°æ®
def load_complete_tang_poetry():
    """åŠ è½½å…¨å”è¯—æ•°æ®"""
    print(" æ­£åœ¨åŠ è½½å…¨å”è¯—æ•°æ®...")
    
    try:
        # å¦‚æœæ‚¨æœ‰å¤„ç†å¥½çš„CSVæ–‡ä»¶
        if os.path.exists('å…¨å”è¯—åˆ†æç»“æœ.csv'):
            complete_df = pd.read_csv('å…¨å”è¯—åˆ†æç»“æœ.csv')
            print(f" ä»CSVæ–‡ä»¶åŠ è½½å…¨å”è¯—æ•°æ®: {len(complete_df)} é¦–")
            return complete_df
        
        # å¦‚æœè¦ä»åŸå§‹JSONæ–‡ä»¶åŠ è½½ï¼ˆä½¿ç”¨æ‚¨ä¹‹å‰çš„ä»£ç ï¼‰
        def load_tang_poetry_in_batches(tang_poetry_folder="å…¨å”è¯—", sample_ratio=0.3):
            print(" å¼€å§‹åˆ†æ‰¹åŠ è½½å…¨å”è¯—æ•°æ®...")
            
            # æ‰¾åˆ°æ‰€æœ‰å”è¯—æ–‡ä»¶
            tang_files = [f for f in os.listdir(tang_poetry_folder)
                         if f.startswith('poet.tang') and f.endswith('.json')]
            
            print(f"æ‰¾åˆ° {len(tang_files)} ä¸ªå”è¯—æ•°æ®æ–‡ä»¶")
            
            all_poems = []
            file_sample_size = max(1, int(len(tang_files) * sample_ratio))
            sampled_files = sorted(tang_files)[:file_sample_size]
            
            print(f" é‡‡æ ·ç­–ç•¥: ä» {len(tang_files)} ä¸ªæ–‡ä»¶ä¸­é€‰å– {file_sample_size} ä¸ªè¿›è¡Œåˆ†æ")
            
            for i, poet_file in enumerate(sampled_files):
                file_path = os.path.join(tang_poetry_folder, poet_file)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        content = f.read()
                        content = content.replace('\"', '"').replace('\"', '"')
                        poems = json.loads(content)
                        
                        # å¦‚æœæ•°æ®é‡å¾ˆå¤§ï¼Œå¯¹æ¯ä¸ªæ–‡ä»¶ä¹Ÿè¿›è¡Œé‡‡æ ·
                        if len(poems) > 1000:
                            sample_size = min(800, len(poems))
                            poems = poems[:sample_size]
                            
                        all_poems.extend(poems)
                        print(f" å·²åŠ è½½ {poet_file}: {len(poems)} é¦–è¯— (æ€»è®¡: {len(all_poems)})")
                        
                except Exception as e:
                    print(f" è¯»å–æ–‡ä»¶ {poet_file} æ—¶å‡ºé”™: {e}")
                    continue
            
            print(f" æœ€ç»ˆåŠ è½½ {len(all_poems)} é¦–å”è¯—è¿›è¡Œåˆ†æ")
            return all_poems

        def process_tang_poetry_data(poems):
            """å¤„ç†å”è¯—æ•°æ®ï¼Œæ·»åŠ æ—¶æœŸå’Œé¢˜æä¿¡æ¯"""
            print(" å¤„ç†è¯—æ­Œæ•°æ®...")
            
            # å®Œæ•´çš„å”æœæ—¶æœŸæ˜ å°„
            period_mapping = {
                # åˆå” (618-712)
                'éª†å®¾ç‹': 'åˆå”', 'é™ˆå­æ˜‚': 'åˆå”', 'æœå®¡è¨€': 'åˆå”', 'ç‹å‹ƒ': 'åˆå”',
                'æ¨ç‚¯': 'åˆå”', 'å¢ç…§é‚»': 'åˆå”', 'æ²ˆä½ºæœŸ': 'åˆå”', 'å®‹ä¹‹é—®': 'åˆå”',
                'è´ºçŸ¥ç« ': 'åˆå”', 'å¼ è‹¥è™š': 'åˆå”', 'å¼ ä¹é¾„': 'åˆå”',
                
                # ç››å” (713-765)
                'æç™½': 'ç››å”', 'æœç”«': 'ç››å”', 'ç‹ç»´': 'ç››å”', 'å­Ÿæµ©ç„¶': 'ç››å”',
                'ç‹æ˜Œé¾„': 'ç››å”', 'å²‘å‚': 'ç››å”', 'é«˜é€‚': 'ç››å”', 'ç‹ä¹‹æ¶£': 'ç››å”',
                'å´”é¢¢': 'ç››å”', 'å‚¨å…‰ç¾²': 'ç››å”', 'å¸¸å»º': 'ç››å”', 'æé¢€': 'ç››å”',
                
                # ä¸­å” (766-835)
                'ç™½å±…æ˜“': 'ä¸­å”', 'å…ƒç¨¹': 'ä¸­å”', 'éŸ©æ„ˆ': 'ä¸­å”', 'æŸ³å®—å…ƒ': 'ä¸­å”',
                'åˆ˜ç¦¹é”¡': 'ä¸­å”', 'å­ŸéƒŠ': 'ä¸­å”', 'è´¾å²›': 'ä¸­å”', 'æè´º': 'ä¸­å”',
                'éŸ¦åº”ç‰©': 'ä¸­å”', 'å¼ ç±': 'ä¸­å”', 'ç‹å»º': 'ä¸­å”', 'å¢çº¶': 'ä¸­å”',
                'æç›Š': 'ä¸­å”', 'å¼ ç¥œ': 'ä¸­å”',
                
                # æ™šå” (836-907)
                'æå•†éš': 'æ™šå”', 'æœç‰§': 'æ™šå”', 'æ¸©åº­ç­ ': 'æ™šå”', 'éŸ¦åº„': 'æ™šå”',
                'è®¸æµ‘': 'æ™šå”', 'çš®æ—¥ä¼‘': 'æ™šå”', 'é™†é¾Ÿè’™': 'æ™šå”', 'ç½—éš': 'æ™šå”',
                'è‚å¤·ä¸­': 'æ™šå”', 'æœè€é¹¤': 'æ™šå”', 'éŸ©å“': 'æ™šå”'
            }
            
            # é¢˜æå…³é”®è¯æ˜ å°„
            genre_keywords = {
                'å±±æ°´': 'å±±æ°´ç”°å›­', 'ç”°å›­': 'å±±æ°´ç”°å›­', 'è‡ªç„¶': 'å±±æ°´ç”°å›­', 'å±±': 'å±±æ°´ç”°å›­', 'æ°´': 'å±±æ°´ç”°å›­',
                'è¾¹å¡': 'è¾¹å¡æˆ˜äº‰', 'æˆ': 'è¾¹å¡æˆ˜äº‰', 'æˆ˜': 'è¾¹å¡æˆ˜äº‰', 'å†›': 'è¾¹å¡æˆ˜äº‰', 'å¡': 'è¾¹å¡æˆ˜äº‰',
                'å’ç‰©': 'å’ç‰©æŠ’æ€€', 'å’': 'å’ç‰©æŠ’æ€€', 'ç‰©': 'å’ç‰©æŠ’æ€€', 'æ¢…': 'å’ç‰©æŠ’æ€€', 'èŠ': 'å’ç‰©æŠ’æ€€',
                'é€åˆ«': 'é€åˆ«æ€€äºº', 'ç¦»åˆ«': 'é€åˆ«æ€€äºº', 'åˆ«': 'é€åˆ«æ€€äºº', 'èµ ': 'é€åˆ«æ€€äºº',
                'æ€€å¤': 'æ€€å¤å’å²', 'å²': 'æ€€å¤å’å²', 'å¤': 'æ€€å¤å’å²', 'å¤äºº': 'æ€€å¤å’å²',
                'çˆ±æƒ…': 'çˆ±æƒ…é—ºæ€¨', 'ç›¸æ€': 'çˆ±æƒ…é—ºæ€¨', 'æ€¨': 'çˆ±æƒ…é—ºæ€¨', 'æƒ…': 'çˆ±æƒ…é—ºæ€¨',
                'æ”¿æ²»': 'æ”¿æ²»æŠ’æ€€', 'è®½': 'æ”¿æ²»æŠ’æ€€', 'è°': 'æ”¿æ²»æŠ’æ€€', 'å›½': 'æ”¿æ²»æŠ’æ€€',
                'éšé€¸': 'éšé€¸é—²é€‚', 'éš': 'éšé€¸é—²é€‚', 'é—²': 'éšé€¸é—²é€‚', 'é€¸': 'éšé€¸é—²é€‚'
            }
            
            processed_poems = []
            
            for i, poem in enumerate(poems):
                if i % 1000 == 0 and i > 0:
                    print(f"  å·²å¤„ç† {i}/{len(poems)} é¦–è¯—...")
                    
                author = poem.get('author', '')
                title = poem.get('title', '')
                paragraphs = poem.get('paragraphs', [])
                
                # åˆå¹¶å†…å®¹
                if isinstance(paragraphs, list):
                    content = ' '.join(paragraphs)
                else:
                    content = str(paragraphs)
                
                # æ¸…ç†å†…å®¹
                content = content.replace(' ', '').replace('ã€€', '')
                
                # è·³è¿‡å†…å®¹å¤ªçŸ­çš„è¯—
                if len(content) < 10:
                    continue
                
                # æ·»åŠ æ—¶æœŸä¿¡æ¯
                period = period_mapping.get(author, 'æœªçŸ¥æ—¶æœŸ')
                
                # æ¨æ–­é¢˜æ
                genre = 'å…¶ä»–'
                search_text = title + content
                for keyword, genre_name in genre_keywords.items():
                    if keyword in search_text:
                        genre = genre_name
                        break
                
                processed_poems.append({
                    'author': author,
                    'title': title,
                    'content': content,
                    'period': period,
                    'genre': genre,
                    'char_count': len(content)
                })
            
            print(f"âœ… æ•°æ®å¤„ç†å®Œæˆ: {len(processed_poems)} é¦–æœ‰æ•ˆè¯—æ­Œ")
            return pd.DataFrame(processed_poems)

        # åŠ è½½æ•°æ®
        all_poems = load_tang_poetry_in_batches("å…¨å”è¯—", sample_ratio=0.3)
        complete_df = process_tang_poetry_data(all_poems)
        
        # ä¿å­˜ä¸ºCSVä»¥ä¾¿ä¸‹æ¬¡ä½¿ç”¨
        complete_df.to_csv('å…¨å”è¯—åˆ†æç»“æœ.csv', index=False, encoding='utf-8-sig')
        print(" å…¨å”è¯—æ•°æ®å·²ä¿å­˜ä¸º 'å…¨å”è¯—åˆ†æç»“æœ.csv'")
        
        return complete_df
        
    except Exception as e:
        print(f" åŠ è½½å…¨å”è¯—æ•°æ®å¤±è´¥: {e}")
        print(" ä½¿ç”¨æ¨¡æ‹Ÿæ•°æ®ç»§ç»­åˆ†æ...")
        return create_sample_complete_data()

def create_sample_three_hundred_data():
    """åˆ›å»ºæ¨¡æ‹Ÿçš„ã€Šå”è¯—ä¸‰ç™¾é¦–ã€‹æ•°æ®"""
    print(" åˆ›å»ºæ¨¡æ‹Ÿçš„ã€Šå”è¯—ä¸‰ç™¾é¦–ã€‹æ•°æ®...")
    
    # æ ¹æ®æ‚¨ä¹‹å‰çš„åˆ†æç»“æœåˆ›å»ºæ•°æ®
    data = {
        'author': [],
        'period': [],
        'genre': []
    }
    
    # æ—¶æœŸåˆ†å¸ƒ
    periods = {
        'ç››å”': 164,
        'æœªçŸ¥æ—¶æœŸ': 105, 
        'ä¸­å”': 44,
        'æ™šå”': 43,
        'åˆå”': 10
    }
    
    # é¢˜æåˆ†å¸ƒ
    genres = {
        'çˆ±æƒ…é—ºæ€¨': 86,
        'å±±æ°´ç”°å›­': 75,
        'æ€€å¤å’å²': 74,
        'å…¶ä»–': 47,
        'è¾¹å¡æˆ˜äº‰': 31,
        'é€åˆ«æ€€äºº': 20,
        'å’ç‰©æŠ’æ€€': 17,
        'æ”¿æ²»æŠ’æ€€': 16
    }
    
    # è¯—äººåˆ†å¸ƒï¼ˆä¸»è¦è¯—äººï¼‰
    poets = {
        'æç™½': 51, 'æœç”«': 41, 'ç‹ç»´': 29, 'æå•†éš': 25, 'å­Ÿæµ©ç„¶': 16,
        'éŸ¦åº”ç‰©': 13, 'åˆ˜é•¿å¿': 11, 'ç‹æ˜Œé¾„': 9, 'æœç‰§': 9, 'æé¢€': 8
    }
    
    # åˆ›å»ºæ•°æ®
    author_list = []
    period_list = []
    genre_list = []
    
    # æ·»åŠ ä¸»è¦è¯—äºº
    for poet, count in poets.items():
        author_list.extend([poet] * count)
    
    # æ·»åŠ å…¶ä»–è¯—äºº
    remaining_poems = 366 - sum(poets.values())
    author_list.extend(['å…¶ä»–è¯—äºº'] * remaining_poems)
    
    # æ·»åŠ æ—¶æœŸ
    for period, count in periods.items():
        period_list.extend([period] * count)
    
    # æ·»åŠ é¢˜æ
    for genre, count in genres.items():
        genre_list.extend([genre] * count)
    
    # ç¡®ä¿æ‰€æœ‰æ•°ç»„é•¿åº¦ç›¸åŒ
    min_length = min(len(author_list), len(period_list), len(genre_list))
    
    data = {
        'author': author_list[:min_length],
        'period': period_list[:min_length], 
        'genre': genre_list[:min_length]
    }
    
    three_hundred_df = pd.DataFrame(data)
    print(f" æ¨¡æ‹Ÿã€Šå”è¯—ä¸‰ç™¾é¦–ã€‹æ•°æ®åˆ›å»ºå®Œæˆ: {len(three_hundred_df)} é¦–")
    return three_hundred_df

def create_sample_complete_data():
    """åˆ›å»ºæ¨¡æ‹Ÿçš„ã€Šå…¨å”è¯—ã€‹æ•°æ®"""
    print(" åˆ›å»ºæ¨¡æ‹Ÿçš„ã€Šå…¨å”è¯—ã€‹æ•°æ®...")
    
    # ä¸»è¦è¯—äººåŠå…¶ä½œå“æ•°é‡
    main_poets = {
        'ç™½å±…æ˜“': 3009, 'æœç”«': 1489, 'æç™½': 1207, 'å…ƒç¨¹': 910, 'åˆ˜ç¦¹é”¡': 886,
        'é½å·±': 826, 'è´¯ä¼‘': 743, 'æ˜“é™': 721, 'æ— åæ°': 684, 'é™†é¾Ÿè’™': 635,
        'æå•†éš': 616, 'éŸ¦åº”ç‰©': 586, 'ç‹å»º': 580, 'å¼ ç¥œ': 565, 'æœç‰§': 544
    }
    
    # æ—¶æœŸåˆ†å¸ƒ
    periods = {
        'æœªçŸ¥æ—¶æœŸ': 47869,
        'ä¸­å”': 5419, 
        'ç››å”': 3029,
        'æ™šå”': 986,
        'åˆå”': 304
    }
    
    # é¢˜æåˆ†å¸ƒ
    genres = {
        'å±±æ°´ç”°å›­': 29675,
        'å…¶ä»–': 15411,
        'å’ç‰©æŠ’æ€€': 5753, 
        'çˆ±æƒ…é—ºæ€¨': 2141,
        'æ€€å¤å’å²': 1882,
        'è¾¹å¡æˆ˜äº‰': 1346,
        'éšé€¸é—²é€‚': 913,
        'é€åˆ«æ€€äºº': 244,
        'æ”¿æ²»æŠ’æ€€': 242
    }
    
    # åˆ›å»ºæ•°æ®
    author_list = []
    period_list = []
    genre_list = []
    
    # æ·»åŠ ä¸»è¦è¯—äºº
    for poet, count in main_poets.items():
        author_list.extend([poet] * count)
    
    # æ·»åŠ å…¶ä»–è¯—äºº
    remaining_poems = 57607 - sum(main_poets.values())
    author_list.extend(['å…¶ä»–è¯—äºº'] * remaining_poems)
    
    # æ·»åŠ æ—¶æœŸ
    for period, count in periods.items():
        period_list.extend([period] * count)
    
    # æ·»åŠ é¢˜æ  
    for genre, count in genres.items():
        genre_list.extend([genre] * count)
    
    # ç¡®ä¿æ‰€æœ‰æ•°ç»„é•¿åº¦ç›¸åŒ
    min_length = min(len(author_list), len(period_list), len(genre_list))
    
    data = {
        'author': author_list[:min_length],
        'period': period_list[:min_length],
        'genre': genre_list[:min_length]
    }
    
    complete_df = pd.DataFrame(data)
    print(f" æ¨¡æ‹Ÿã€Šå…¨å”è¯—ã€‹æ•°æ®åˆ›å»ºå®Œæˆ: {len(complete_df)} é¦–")
    return complete_df

# 2. é€‰æœ¬åè§å¯è§†åŒ–
def plot_selection_bias(three_hundred_df, complete_df):
    """ç»˜åˆ¶é€‰æœ¬åè§å¯¹æ¯”å›¾"""
    print(" ç”Ÿæˆé€‰æœ¬åè§å¯¹æ¯”å›¾...")
    
    # è®¡ç®—æ—¶æœŸåˆ†å¸ƒæ¯”ä¾‹
    period_300 = three_hundred_df['period'].value_counts(normalize=True)
    period_complete = complete_df['period'].value_counts(normalize=True)
    
    # ç¡®ä¿ä¸¤ä¸ªæ•°æ®é›†çš„æ—¶æœŸé¡ºåºä¸€è‡´
    all_periods = sorted(set(period_300.index) | set(period_complete.index))
    period_300 = period_300.reindex(all_periods, fill_value=0)
    period_complete = period_complete.reindex(all_periods, fill_value=0)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # æ—¶æœŸåˆ†å¸ƒå¯¹æ¯”
    x = np.arange(len(all_periods))
    width = 0.35
    
    ax1.bar(x - width/2, period_300.values, width, label='ã€Šä¸‰ç™¾é¦–ã€‹', alpha=0.8, color='#FF6B6B')
    ax1.bar(x + width/2, period_complete.values, width, label='ã€Šå…¨å”è¯—ã€‹', alpha=0.8, color='#4ECDC4')
    ax1.set_title('æ—¶æœŸåˆ†å¸ƒå¯¹æ¯” - é€‰æœ¬åè§æ˜æ˜¾', fontsize=14, fontweight='bold')
    ax1.set_ylabel('æ¯”ä¾‹')
    ax1.set_xticks(x)
    ax1.set_xticklabels(all_periods, rotation=45)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # è®¡ç®—åè§ç³»æ•°
    bias_ratios = {}
    for period in all_periods:
        if period_complete[period] > 0:
            bias_ratios[period] = period_300[period] / period_complete[period]
        else:
            bias_ratios[period] = 0
    
    colors = ['red' if ratio > 1 else 'blue' for ratio in bias_ratios.values()]
    bars = ax2.bar(bias_ratios.keys(), bias_ratios.values(), color=colors, alpha=0.7)
    ax2.set_title('é€‰æœ¬åè§ç³»æ•° (ä¸‰ç™¾é¦–/å…¨å”è¯—)', fontsize=14, fontweight='bold')
    ax2.set_ylabel('åè§å€æ•°')
    ax2.axhline(y=1, color='black', linestyle='--', alpha=0.5)
    ax2.text(0.5, 1.05, 'æ— åè§çº¿', transform=ax2.get_xaxis_transform(), 
             ha='center', fontsize=10)
    plt.xticks(rotation=45)
    
    # åœ¨æŸ±å­ä¸Šæ·»åŠ æ•°å€¼
    for bar, ratio in zip(bars, bias_ratios.values()):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{ratio:.1f}x', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig('é€‰æœ¬åè§åˆ†æ.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(" é€‰æœ¬åè§ç³»æ•°:")
    for period, ratio in bias_ratios.items():
        if ratio > 0:
            print(f"  {period}: {ratio:.1f}å€")
    
    return bias_ratios

# 3. é¢˜ææµå˜æ—¶é—´çº¿
def plot_genre_evolution_timeline(complete_df):
    """ç»˜åˆ¶é¢˜æéšæ—¶é—´æ¼”å˜çš„æ—¶é—´çº¿å›¾"""
    print(" ç”Ÿæˆé¢˜ææµå˜æ—¶é—´çº¿å›¾...")
    
    # æŒ‰æ—¶æœŸå’Œé¢˜æåˆ†ç»„
    period_genre = complete_df.groupby(['period', 'genre']).size().unstack(fill_value=0)
    
    # åªå–ä¸»è¦çš„æ—¶æœŸ
    main_periods = ['åˆå”', 'ç››å”', 'ä¸­å”', 'æ™šå”']
    period_genre = period_genre.reindex(main_periods, fill_value=0)
    
    # è®¡ç®—æ¯ä¸ªæ—¶æœŸå†…çš„é¢˜ææ¯”ä¾‹
    period_genre_pct = period_genre.div(period_genre.sum(axis=1), axis=0)
    
    # åªå–å‰6ä¸ªä¸»è¦é¢˜æ
    top_genres = period_genre_pct.sum().nlargest(6).index
    period_genre_pct = period_genre_pct[top_genres]
    
    plt.figure(figsize=(14, 8))
    
    # è®¾ç½®é¢œè‰²
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFD166', '#96CEB4', '#6A0572']
    
    bottom_vals = np.zeros(len(main_periods))
    for i, genre in enumerate(period_genre_pct.columns):
        values = [period_genre_pct.loc[p, genre] if p in period_genre_pct.index else 0 for p in main_periods]
        plt.bar(main_periods, values, bottom=bottom_vals, 
                label=genre, color=colors[i], alpha=0.8, edgecolor='white')
        bottom_vals += values
    
    plt.title('å…¨å”è¯—é¢˜ææµå˜æ—¶é—´çº¿åˆ†æ', fontsize=16, fontweight='bold')
    plt.ylabel('é¢˜ææ¯”ä¾‹', fontsize=12)
    plt.xlabel('å†å²æ—¶æœŸ', fontsize=12)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='é¢˜æç±»å‹')
    plt.grid(True, alpha=0.3, axis='y')
    
    # æ·»åŠ æ¯”ä¾‹æ ‡æ³¨
    for i, period in enumerate(main_periods):
        bottom = 0
        for genre in period_genre_pct.columns:
            value = period_genre_pct.loc[period, genre] if period in period_genre_pct.index else 0
            if value > 0.05:  # åªæ ‡æ³¨å¤§äº5%çš„éƒ¨åˆ†
                plt.text(i, bottom + value/2, f'{value:.1%}', 
                        ha='center', va='center', fontsize=9, fontweight='bold',
                        color='white' if value > 0.2 else 'black')
            bottom += value
    
    plt.tight_layout()
    plt.savefig('å…¨å”è¯—é¢˜ææµå˜æ—¶é—´çº¿.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(" é¢˜ææµå˜æ—¶é—´çº¿å›¾ç”Ÿæˆå®Œæˆ")

# 4. è¯—äººå½±å“åŠ›ç½‘ç»œå›¾
def plot_poet_network(three_hundred_df, complete_df):
    """ç»˜åˆ¶è¯—äººå½±å“åŠ›ç½‘ç»œå›¾"""
    print(" ç”Ÿæˆè¯—äººå½±å“åŠ›ç½‘ç»œå›¾...")
    
    # è®¡ç®—è¯—äººåœ¨ä¸¤ä¸ªæ•°æ®é›†ä¸­çš„é‡è¦æ€§å¾—åˆ†
    poet_importance = {}
    
    all_poets = set(three_hundred_df['author']).union(set(complete_df['author']))
    
    for poet in all_poets:
        # åœ¨ã€Šä¸‰ç™¾é¦–ã€‹ä¸­çš„æ¯”ä¾‹
        count_300 = three_hundred_df[three_hundred_df['author'] == poet].shape[0]
        score_300 = count_300 / len(three_hundred_df) if len(three_hundred_df) > 0 else 0
        
        # åœ¨ã€Šå…¨å”è¯—ã€‹ä¸­çš„æ¯”ä¾‹  
        count_complete = complete_df[complete_df['author'] == poet].shape[0]
        score_complete = count_complete / len(complete_df) if len(complete_df) > 0 else 0
        
        # é‡è¦æ€§ = é€‰æœ¬ä»£è¡¨æ€§ Ã— å®é™…äº§é‡ (æ”¾å¤§æ˜¾ç¤º)
        importance = score_300 * score_complete * 1000000  # æ”¾å¤§å€æ•°
        
        if importance > 0 and count_300 + count_complete > 5:  # è¿‡æ»¤æ‰å¤ªå°çš„è¯—äºº
            poet_importance[poet] = {
                'importance': importance,
                'count_300': count_300,
                'count_complete': count_complete,
                'score_300': score_300,
                'score_complete': score_complete
            }
    
    # å–å‰15ä½é‡è¦è¯—äºº
    top_poets = dict(sorted(poet_importance.items(), 
                           key=lambda x: x[1]['importance'], reverse=True)[:15])
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))
    
    # æ•£ç‚¹å›¾ï¼šé€‰æœ¬ä»£è¡¨æ€§ vs å®é™…äº§é‡
    poets = list(top_poets.keys())
    scores_300 = [top_poets[p]['score_300'] for p in poets]
    scores_complete = [top_poets[p]['score_complete'] for p in poets]
    importance = [top_poets[p]['importance'] for p in poets]
    
    scatter = ax1.scatter(scores_300, scores_complete, s=importance, 
                         alpha=0.6, c=importance, cmap='viridis')
    
    for i, poet in enumerate(poets):
        ax1.annotate(poet, (scores_300[i], scores_complete[i]), 
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=9, alpha=0.8, fontweight='bold')
    
    ax1.set_xlabel('ã€Šä¸‰ç™¾é¦–ã€‹ä¸­çš„æ¯”ä¾‹ (é€‰æœ¬ä»£è¡¨æ€§)', fontsize=12)
    ax1.set_ylabel('ã€Šå…¨å”è¯—ã€‹ä¸­çš„æ¯”ä¾‹ (å®é™…äº§é‡)', fontsize=12)
    ax1.set_title('è¯—äººå½±å“åŠ›åˆ†æ: é€‰æœ¬ä»£è¡¨æ€§ vs å®é™…äº§é‡', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    
    # æ·»åŠ é¢œè‰²æ¡
    cbar = plt.colorbar(scatter, ax=ax1)
    cbar.set_label('ç»¼åˆå½±å“åŠ›å¾—åˆ†', rotation=270, labelpad=15)
    
    # æŸ±çŠ¶å›¾ï¼šè¯¦ç»†æ•°æ®å¯¹æ¯”
    x_pos = np.arange(len(poets))
    width = 0.35
    
    counts_300 = [top_poets[p]['count_300'] for p in poets]
    counts_complete = [top_poets[p]['count_complete'] for p in poets]
    
    ax2.bar(x_pos - width/2, counts_300, width, label='ã€Šä¸‰ç™¾é¦–ã€‹', alpha=0.8, color='#FF6B6B')
    ax2.bar(x_pos + width/2, counts_complete, width, label='ã€Šå…¨å”è¯—ã€‹', alpha=0.8, color='#4ECDC4')
    
    ax2.set_xlabel('è¯—äºº', fontsize=12)
    ax2.set_ylabel('ä½œå“æ•°é‡', fontsize=12)
    ax2.set_title('è¯—äººä½œå“æ•°é‡å¯¹æ¯”', fontsize=14, fontweight='bold')
    ax2.set_xticks(x_pos)
    ax2.set_xticklabels(poets, rotation=45, ha='right')
    ax2.legend()
    ax2.grid(True, alpha=0.3, axis='y')
    
    # åœ¨æŸ±å­ä¸Šæ·»åŠ æ•°é‡æ ‡æ³¨
    for i, (count_300, count_complete) in enumerate(zip(counts_300, counts_complete)):
        if count_300 > 0:
            ax2.text(i - width/2, count_300 + max(counts_complete)*0.01, 
                    str(count_300), ha='center', va='bottom', fontsize=8)
        if count_complete > 0:
            ax2.text(i + width/2, count_complete + max(counts_complete)*0.01, 
                    str(count_complete), ha='center', va='bottom', fontsize=8)
    
    plt.tight_layout()
    plt.savefig('è¯—äººå½±å“åŠ›ç½‘ç»œ.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("ğŸ“‹ é‡è¦è¯—äººåˆ†æ:")
    for poet, data in list(top_poets.items())[:10]:
        print(f"  {poet}: ä¸‰ç™¾é¦–{data['count_300']}é¦–, å…¨å”è¯—{data['count_complete']}é¦–, "
              f"å½±å“åŠ›å¾—åˆ†{data['importance']:.0f}")

# ä¸»ç¨‹åº
def main():
    print("=" * 60)
    print(" å”è¯—é€‰æœ¬ä¸å…¨é›†å¯¹æ¯”åˆ†æ")
    print("=" * 60)
    
    # åŠ è½½æ•°æ®
    print("\n1. åŠ è½½æ•°æ®...")
    three_hundred_df = create_sample_three_hundred_data()
    complete_df = load_complete_tang_poetry()
    
    print(f"\n æ•°æ®æ¦‚å†µ:")
    print(f"  ã€Šå”è¯—ä¸‰ç™¾é¦–ã€‹: {len(three_hundred_df)} é¦–è¯—æ­Œ")
    print(f"  ã€Šå…¨å”è¯—ã€‹: {len(complete_df)} é¦–è¯—æ­Œ")
    
    # æ‰§è¡Œåˆ†æ
    print("\n2. å¼€å§‹ç”Ÿæˆåˆ†æå›¾è¡¨...")
    
    # é€‰æœ¬åè§åˆ†æ
    print("\n åˆ†æ1: é€‰æœ¬åè§åˆ†æ")
    bias_ratios = plot_selection_bias(three_hundred_df, complete_df)
    
    # é¢˜ææµå˜åˆ†æ
    print("\n åˆ†æ2: é¢˜ææµå˜æ—¶é—´çº¿")
    plot_genre_evolution_timeline(complete_df)
    
    # è¯—äººå½±å“åŠ›åˆ†æ
    print("\n åˆ†æ3: è¯—äººå½±å“åŠ›ç½‘ç»œ")
    plot_poet_network(three_hundred_df, complete_df)
    
    print("\n" + "=" * 60)
    print(" æ‰€æœ‰åˆ†æå®Œæˆï¼")
    print(" ç”Ÿæˆçš„æ–‡ä»¶:")
    print("  - é€‰æœ¬åè§åˆ†æ.png")
    print("  - å…¨å”è¯—é¢˜ææµå˜æ—¶é—´çº¿.png") 
    print("  - è¯—äººå½±å“åŠ›ç½‘ç»œ.png")
    print("=" * 60)

# è¿è¡Œç¨‹åº
if __name__ == "__main__":
    main()
```
